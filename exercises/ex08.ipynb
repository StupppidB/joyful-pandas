```python
import numpy as np
import pandas as pd
```

```python
earth = pd.read_csv('../data/Earthquake.csv')
```

```python
earth.head()
```

```python
earth['深度等级'] = pd.cut(earth['深度'], [-1e-10,5,10,15,20,30,50,np.inf], labels=['Ⅰ','Ⅱ','Ⅲ','Ⅳ','Ⅴ','Ⅵ','Ⅶ'])
```

```python
earth.sort_values('深度等级')
```

```python
earth['烈度等级'] = pd.cut(earth['烈度'], [-1e-10,3,4,5,np.inf], labels=['Ⅰ','Ⅱ','Ⅲ','Ⅳ'])
```

```python
earth.head()
```

```python
earth.set_index(['深度等级', '烈度等级']).sort_index()
```
