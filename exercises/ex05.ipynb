```python
Course1 = pd.read_csv('../data/Course1.csv')
```

```python
Course2 = pd.read_csv('../data/Course2.csv')
```

```python
Course1.head()
```

```python
Course2.head()
```

```python
Course1_special = Course1.query('课程类别 in ["学科基础课","专业必修课","专业选修课"]')
```

```python
Course1_common = Course1.query('课程类别 not in ["学科基础课","专业必修课","专业选修课"]')
```

```python
Course2_special = Course2.query('课程类别 in ["学科基础课","专业必修课","专业选修课"]')
```

```python
Course2_common = Course2.query('课程类别 not in ["学科基础课","专业必修课","专业选修课"]')
```

```python
special = pd.concat([Course1_special, Course2_special])
```

```python
common = pd.concat([Course1_common, Course2_common])
```

```python
special
```

```python
common
```

```python
all = pd.concat([Course1, Course2])
```

```python
special2 = all.query('课程类别 in ["学科基础课","专业必修课","专业选修课"]')
```

```python
common2 = all.query('课程类别 not in ["学科基础课","专业必修课","专业选修课"]')
```

```python
(special2.equals(special),common2.equals(common))
```
