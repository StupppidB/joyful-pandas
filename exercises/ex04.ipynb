```python
import numpy as np
import pandas as pd
```

### 练习一

```python
drugs = pd.read_csv('../data/Drugs.csv')
```

```python
drugs
```

```python
drugs.rename_axis()
```

```python
df = pd.pivot_table(drugs,index=['State', 'COUNTY', 'SubstanceName'],columns='YYYY',values='DrugReports', fill_value='-').reset_index()
```

```python
df.info()
```

```python
df_ori = df.melt(id_vars=['State', 'COUNTY', 'SubstanceName'],var_name = 'YYYY',value_name='DrugReports').query("DrugReports!='-'").sort_values(['YYYY', 'State', 'COUNTY', 'SubstanceName']).reset_index(drop=True)[['YYYY', 'State', 'COUNTY', 'SubstanceName', 'DrugReports']]
```

```python
df_ori[['YYYY', 'DrugReports']] = df_ori[['YYYY', 'DrugReports']].astype('int64')
```

```python
df_ori
```

```python
df_ori.equals(drugs.sort_values(['YYYY', 'State', 'COUNTY', 'SubstanceName']).reset_index(drop=True))
```

### 练习二

```python
earth = pd.read_csv('../data/Earthquake.csv')
```

```python
earth
```

```python
earth_pivot = pd.pivot_table(earth,index=['日期', '时间', '维度', '经度'],columns='方向',values=['深度', '烈度', '烈度'], fill_value='-').stack(level=0).rename(index={None:'地震参数'})
```

```python
earth_pivot
```

```python
earth_ori = earth_pivot.unstack().stack(0).query("深度!='-' & 烈度!='-' & 距离 !='-'").sort_values(['日期', '时间', '维度', '经度', '方向','距离', '深度', '烈度']).reset_index()[['日期', '时间', '维度', '经度', '方向','距离', '深度', '烈度']]
```

```python
earth_ori[['距离', '深度', '烈度']] = earth_ori[['距离', '深度', '烈度']].astype('float64')
```

```python
earth_ori
```

```python
earth_ori.equals(earth.sort_values(['日期', '时间', '维度', '经度', '方向','距离', '深度', '烈度']).reset_index(drop=True))
```
