### 练习一

```python
thrones = pd.read_csv('../data/Game_of_Thrones_Script.csv')
```

```python
thrones
```

```python
thrones.Name.nunique()
```

```python
thrones.Name.value_counts().head(1)
```

```python
thrones['length'] = thrones.Sentence.apply(lambda x:len(x.split(' ')))
```

```python
thrones.groupby('Name').length.sum().sort_values(ascending=False).head(1)
```
